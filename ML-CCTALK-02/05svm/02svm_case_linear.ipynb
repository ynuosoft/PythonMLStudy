# 支持向量机介绍：核函数：linear

## 1 引用

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import svm
from sklearn import datasets
from sklearn import metrics
```

## 2 数据准备

```python
X01, y01 = datasets.make_classification(n_features=4, random_state=0,n_samples=1000)
```

```python
X01[:5]
```

```python
y01[:5]
```

### 2.1 只提取前面2个特征做为训练数据

```python
X01_2column=X01[:,0:2]
```

```python
X01_2column[:5]
```

### 2.1 数据原始分布

```python
plt.figure(figsize=(14,10))
plt.scatter(x=X01_2column[:,0],y=X01_2column[:,1],c=y01)
plt.show()
```

## 3 准备模型 :线性核函数

```python
clf01 =svm.SVC(kernel="linear")
```

```python
clf01
```

## 4 模型训练 

### 4.1 只提取前面2个特征做为训练数据

```python
%%time
clf01.fit(X01_2column, y01)
```

## 5 模型系数

### 5.1 线性系数

```python
coef01=clf01.coef_
coef01
```

```python
intercept01=clf01.intercept_
intercept01
```

### 5.2 支持向量坐标

```python
svc_vectors01=clf01.support_vectors_
svc_vectors01[:5],svc_vectors01.shape
```

```python
clf01.n_support_
```

```python
#支持向量的坐标索引
clf01.support_[:5]
```

```python
# clf01.dual_coef_[:2]
```

## 6 模型预测

```python
%%time
predict01= clf01.predict(X01_2column)
```

```python
predict01[:20]
```

## 7 结果对比

```python
y01[::50]
```

### 7.1 准确率

```python
print '模型准确率：%f %s' % (metrics.accuracy_score(y_true=y01,y_pred=predict01)*100,'%')
```

## 8 支持向量图示

### 8.1 计算线性分割坐标

```python
min01=X01_2column[:,0].min()
max01=X01_2column[:,0].max()
min01,max01
```

```python
#线性分割线
xcoord01=np.linspace(start=min01.min(),stop=max01,num=100)
```

```python
xcoord01[:5]
```

```python
Y02=((xcoord01*-coef01[0,0]-intercept01)/coef01[0,1])
```

```python
Y02[:5]
```

### 8.2 计算支持向量分割线 以及距离

```python
w01=coef01[0]
a01=-w01[0]/w01[1]
a01
```

#### 8.2.2 线性分割线：中

```python
yy01=a01*xcoord01-(intercept01[0])/w01[1]
yy01[:5]
```

```python
margin01 = 1 / np.sqrt(np.sum(coef01 ** 2))
margin01
```

#### 8.2.3 支持向量分割线：下

```python
yy_down01 = yy01 - np.sqrt(1 + a01 ** 2) * margin01
yy_down01[:5]
```

#### 8.2.4 支持向量分割线：上

```python
yy_up01 = yy01 + np.sqrt(1 + a01 ** 2) * margin01
yy_up01[:5]
```

## 9 画图

### 9.1 红色的点代表是支持向量SV

```python
plt.figure(figsize=(16,12))
#原始数据点
plt.scatter(x=X01[:,0],y=X01[:,1],c=y01)
#支持向量
plt.scatter(x=svc_vectors01[:,0],y=svc_vectors01[:,1],s=80,
                facecolors='r', zorder=10, edgecolors='k')
#支持向量分割线 
plt.plot(xcoord01,yy01,"k--",c='r')
#分割区间
plt.plot(xcoord01,yy_up01,"k--",c='b')
#分割区间
plt.plot(xcoord01,yy_down01,"k--",c='g')
plt.show()
```
