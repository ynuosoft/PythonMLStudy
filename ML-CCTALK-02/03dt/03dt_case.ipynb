```python
import numpy as np
import matplotlib.pyplot as plt
# import graphviz
# print '模型准确率：%f %s' % (metrics.accuracy_score(y_true=y01,y_pred=predict01)*100,'%')
```

```python
from sklearn import tree
from sklearn import datasets
from sklearn import  metrics
```

## 2 准备数据

```python
# 生成随机数据
X01, y01 = datasets.make_blobs(n_samples=1000, centers=6, random_state=40)
```

```python
X01[:5],y01[:5]
```

## 3 数据分布

```python
plt.figure(figsize=[14,10])
plt.scatter(x=X01[:,0],y=X01[:,1],c=y01)
plt.grid()
plt.show()
```

## 4 创建模型

```python
clf01=tree.DecisionTreeClassifier()
```

```python
clf01
```

### 4.1 试试不同的参数

## 5 训练模型

```python
%%time
clf01.fit(X=X01,y=y01)
```

## 6 预测

```python
%%time
predict01=clf01.predict(X=X01)
predict01
```

```python
predict01[:20]
```

## 6.1与原有数据对比

```python
y01[:20]
```

###  6.2准确率 

```python
print '模型准确率：%f %s' % (metrics.accuracy_score(y_true=y01,y_pred=predict01)*100,'%')
```

## 7 模型可视化

```python
import graphviz 
dot_data = tree.export_graphviz(clf01, out_file=None,   
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```
