# 近邻算法练习：KNN

## 1 引用包

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import time
```

```python
from sklearn.neighbors import classification
from sklearn import datasets
from sklearn import metrics
```

## 2 数据加载

```python
data01=datasets.load_iris()
data01.data.shape
```

```python
data01.data[:5]
```

### 2.1 只取2个特征值：图示效果好展示

```python
#目标数据
x01=data01.data[:,:2]
x01[:5]
```

```python
x01.shape
```

```python
x01.dtype
```

```python
#目标分类
y01=data01.target
```

```python
y01
```

### 2.3 原始数据分布

```python
plt.figure(figsize=(14,10))
plt.scatter(x=x01[:,0],y=x01[:,1],c=y01)
plt.show()
```

## 3 模型：创建分类器

```python
classify01=classification.KNeighborsClassifier(n_neighbors=7,weights="distance")
classify01
```

## 4 训练分类器

```python
%%time
classify01.fit(X=x01,y=y01)
```

```python
classify01
```

## 5 预测分类结果

```python
%%time
y_predict01=classify01.predict(X=x01)
```

```python
y_predict01[:20]
```

```python
y01[:20]
```

## 6 结果对比

### 6.1 准确率

```python
#  np.mean(y01==y_predict01)
```

```python
# (y01==y_predict01).sum()
```

```python
metrics.accuracy_score(y_true=y01,y_pred=y_predict01)
```

```python
print '模型准确率：%f %s' % (metrics.accuracy_score(y_true=y01,y_pred=y_predict01)*100,'%')
```

## 7 画图：数据分布

### 7.1 计算画图所有网格点

```python
x_min = x01[:, 0].min() - 1
x_max = x01[:, 0].max() + 1
x_min,x_max
```

```python
y_min = x01[:, 1].min() - 1 
y_max = x01[:, 1].max() + 1
y_min,y_max
```

```python
h = 0.02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
xx[:5]
```

```python
yy[:5]
```

```python
xx.shape,yy.shape
```

```python
%%time
X02_C=np.c_[xx.ravel(),yy.ravel()]
```

```python
X02_C[:2]
```

```python
X02_C.shape
```

#### 7.1.1所有网格数据点预测分类

```python
%%time
z_predict01=classify01.predict(X=X02_C)
```

```python
z_predict01[:20]
```

```python
z_predict01_reshape=z_predict01.reshape(xx.shape)
```

```python
z_predict01_reshape[:5]
```

### 7.2 图示分类效果

```python
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
plt.figure(figsize=(14,10))
#区块划分
# plt.pcolormesh(xx,yy,z_predict01_reshape,cmap=cmap_light)
plt.pcolormesh(xx,yy,z_predict01_reshape)
#数据点
plt.scatter(x=x01[:,0],y=x01[:,1],c=y01,cmap=cmap_bold)
#
plt.show()
```
