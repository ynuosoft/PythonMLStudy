# 画图：numpy矩阵+函数

## 1 与numpy矩阵

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
N01 =50
x01 = np.random.rand(N01)
y01 = np.random.rand(N01)
x01,y01
```

```python
plt.figure(figsize=[14,10])
plt.scatter(x01,y01,c=y01)
plt.xlabel("numpy")
plt.show()
```

## 2.1 函数

```python
x02=x01*np.pi**2
x02
```

```python
plt.figure(figsize=[14,10])
plt.scatter(x02,y01,c=x02)
plt.xlabel("function")
plt.show()
```

### 2.2 颜色随机，大小随机

```python
colors01 = np.random.rand(N01)
area01 = np.pi * (15 * np.random.rand(N01))**2
plt.figure(figsize=[14,10])
plt.scatter(x01, y01, s=area01, c=colors01, alpha=0.5)
plt.show()
```
