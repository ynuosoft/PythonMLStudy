# 画图：布局： subplot

## 1 引用

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

## 2 生成数据

```python
x01=np.arange(100).reshape(4,-1)
y01=np.linspace(start=1,stop=1000,num=100).reshape(4,-1)
```

```python
y01[0],x01[0]
```

## 3 画图布局

```python
plt.figure(1,figsize=[14,18])
#图1
plt.subplot(2,2,1)
plt.scatter(x01[0],y01[0])
plt.xlabel("pic 1")
plt.title("pic 1 title")
#图2
plt.subplot(2,2,2)
plt.plot(x01[1],y01[1])
plt.xlabel("pic 2")
#图3
plt.subplot(2,2,3)
plt.plot(x01[2],y01[2])
plt.xlabel("pic 3")
#图4
plt.subplot(2,2,4)
plt.scatter(x01[3],y01[3])
plt.xlabel("pic 3")
#显示
plt.show()
```

### 4.1 一个画板，一个坐标，多个线图

```python
fig01,ax01 = plt.subplots(figsize=[15,10])

#线条1
plt.plot(x01[0],y01[0],label="line 1111111")

#线条2
plt.plot(y01[0],x01[0],label="line 22222222")

#线条3
line3,=ax01.plot(x01[1],y01[1],label="line 333333")
line3.set_dashes([10,20,30,40])

#线条4
line4,=ax01.plot(y01[1],x01[1],label="line 44444444",dashes=[20,10,30,5])

#设置legend
ax01.legend(loc='lower right')

#增加网格
plt.grid()
plt.show()
```

### 4.2 一个画板，一个坐标系， 多个线图 函数线图

```python
x = np.linspace(0, 10, 500)
dashes = [10, 5, 100, 5]

fig, ax = plt.subplots(figsize=[14,12])
line1, = ax.plot(x, np.sin(x), '--', linewidth=2,
                 label='line 11111111')
line1.set_dashes(dashes)

line2, = ax.plot(x, -1 * np.sin(x), dashes=[30, 5, 10, 5],
                 label='line 222222222')

ax.legend(loc='lower right')
plt.show()
```
