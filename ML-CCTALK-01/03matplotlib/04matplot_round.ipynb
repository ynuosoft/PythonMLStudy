# 画图：画圆：round

## 1 引用

```python
import numpy as np
import matplotlib.pyplot as plt  
```

## 2 定义画圆函数

```python
"""
画圆函数
"""
def round_01(radius=10,center=(0,0),semidot=100):
    #基于 center(0,0)坐标系
    x01=np.linspace(start=0-radius,stop=radius,num=semidot)
    y01=np.sqrt(radius**2-x01**2)
    #转换为真实坐标系
    x01_real=x01+center[0]
    y01_real=y01+center[1]
    y02_real=(y01*-1)+center[1]
    #画图
    plt.figure(figsize=[14,14])
    #中心点
    plt.scatter(center[0],center[1],linewidths=10,s=50)
    #圆
    plt.plot(x01_real,y01_real,'--')
    plt.plot(x01_real,y02_real,'-')
    #画半径线
    
    semidot_01=int(semidot/3)
    x01_real_semi=x01_real[::semidot_01]
    y01_real_semi=y01_real[::semidot_01]
    y02_real_semi=y02_real[::semidot_01]
    for intSemi in range(len(x01_real_semi)):
        plt.plot([x01_real_semi[intSemi],center[0]],[y01_real_semi[intSemi],center[1]],'--')
        plt.plot([x01_real_semi[intSemi],center[0]],[y02_real_semi[intSemi],center[1]],'')
    plt.title("round")
    plt.show()
    return
```

## 3 调用函数

```python
round_01(500,(-1000,-1000),5000)
```
