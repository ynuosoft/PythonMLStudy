# 画图 ：基本点、线、面

## 1 引用

```python
import matplotlib.pyplot as plt
```

### 1.1画点 采用plot属性

```python
x01=[1,2,3,4,5]
y01=[4,8,16,32,64]
```

```python
#画图
plt.figure(figsize=[14,10])
plt.plot(x01, y01, 'ro')
#设置坐标
# plt.axis([0, 10, 0, 70])
plt.xlabel(u"dot")
plt.show()
```

## 1.2 画点 采用scatter

```python
plt.figure(figsize=[14,10])
plt.scatter(y01,x01,s=130)
plt.scatter(x01,y01)
plt.axis([0, 80, 0, 80])
plt.xlabel("dot 2")
plt.show()
```

## 1.3 画线+画面

```python
plt.figure(figsize=[14,10])
plt.plot(x01,y01)
plt.axis([0, 7, 0, 70])

#画面
X02=[1,5,6,1]
y02=[20,9,57,20]
plt.plot(X02,y02,'--')
plt.xlabel(u"line")
plt.show()
```
