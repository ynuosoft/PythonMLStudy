# 矩阵与画图

## 1 高斯分布

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 2, 0.5
v = np.random.normal(mu,sigma,10000)
```

```python
v[:20]
```

```python
plt.figure(figsize=(12,10))
plt.hist(v, bins=50, normed=1) 
plt.show()
```

## NP函数

```python
(n, bins) = np.histogram(v, bins=50, normed=True)
```

```python
n
```

```python
bins
```

```python
plt.figure(figsize=(12,10))
plt.plot(.5*(bins[1:]+bins[:-1]), n)
plt.show()
```
